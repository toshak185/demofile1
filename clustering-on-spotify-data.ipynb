```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
df1['artist_name']=lb.fit_transform(df['artist_name'])
df1['track_name']=lb.fit_transform(df['track_name'])
df1.head()
```

```python
plt.figure(figsize=(20,25), facecolor='white')# defining  canvas size
plotnumber = 1 # initializing plotnumber variable to 1 it will maintain the count of how many graph is going to plot in canvas 


for column in df1: # iteration of columns / acessing the columns from  dataset 
    if plotnumber<=10 :     
        ax = plt.subplot(2,5,plotnumber)  
        sns.distplot(x=df1[column]) ## it give frequency hue =value counts
        plt.xlabel(column,fontsize=20) #assigning name to x-axis and "name" font size is 20
        
    plotnumber+=1 
plt.show()# to show graph
```

```python
#Bivariate analysis
plt.figure(figsize=(20,25), facecolor='white')# defining  canvas size
plotnumber = 1 # initializing plotnumber variable to 1 it will maintain the count of how many graph is going to plot in canvas 


for column in df1.columns: # iteration of columns / acessing the columns from  dataset 
    if plotnumber<=10 :     # as there are 9 columns in the data
        ax = plt.subplot(2,5,plotnumber)  
        sns.histplot(x=df1[column],hue=df1.popularity) ## it give frequency hue =value counts
        plt.xlabel(column,fontsize=20) #assigning name to x-axis and "name" font size is 20
        plt.ylabel('popularity',fontsize=20)#assigning name to y-axis and "name" font size is 20
    plotnumber+=1 # increment of plotnumber 
plt.show()# to show graph
```

```python
#checking for outliers
# let's see how data is distributed for every column
plt.figure(figsize=(20,25), facecolor='white')
plotnumber = 1#counter

for column in df1:#columns form data1
    if plotnumber<=10 :#checking whether plot number is less or equal to 10
        ax = plt.subplot(5,2,plotnumber)# 
        sns.boxplot(df1[column])# ploting boxplot fo outlier 
        plt.xlabel(column,fontsize=20)
        
    plotnumber+=1
plt.show()
```

```python
#Elbow plot
sse = []
k_rng = range(1,10)
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df1[['artist_name','track_name','popularity','year','danceability','energy','mode','acousticness','liveness','tempo','cluster']])
    sse.append(km.inertia_)
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse)
```

From above elbow diagram we can see the optimal value of K will be K=3.
